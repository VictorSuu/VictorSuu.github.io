```python
import numpy as np
import open3d as o3d
import plotly.graph_objects as go
import plotly.figure_factory as ff
import plotly.io as pio
```

```python
def sphere(xyz, radius=1):
    mesh = o3d.geometry.TriangleMesh.create_sphere(radius=radius)
    mesh.translate([xyz[0], xyz[1], xyz[2]])

    vertices = np.asarray(mesh.vertices)
    triangles = np.asarray(mesh.triangles)
    x = vertices[:, 0]
    y = vertices[:, 1]
    z = vertices[:, 2]
    
    return x, y, z, i, j, k
```

```python
x, y, z, i, j, k = sphere(np.zeros(3), radius=1)
data = [go.Mesh3d(x=x, y=y, z=z,
                  i=i, j=j, k=k,
                  hovertemplate="x: %{x}<br>y: %{y}<br>z: %{z}<extra></extra>",
                  hoverlabel=dict(bgcolor="white"),
                  colorscale="algae",
                  intensity=z,
                  showscale=False)]
x, y, z, i, j, k = sphere([0.71, 0.41, 0.57], radius=0.02)
data.append(go.Mesh3d(x=x, y=y, z=z,
                      i=i, j=j, k=k,
                      hovertemplate="<b>x: 0.71<br>y: 0.41<br>z: 0.57</b><extra></extra>",
                      hoverlabel=dict(bgcolor="white"),
                      color="white",
                      showscale=False))
fig = go.Figure(data)

camera = dict(eye=dict(x=0.7, y=0.7, z=1.2),
              up=dict(x=0, y=1, z=0))
fig.update_layout(scene=dict(
                    xaxis=dict(visible=False),
                    yaxis=dict(visible=False),
                    zaxis=dict(visible=False),
                    aspectmode='data'),
                  height=700,
                  margin=dict(r=0, l=0, b=0, t=0, pad=0),
                  scene_camera=camera,
                  scene_dragmode="orbit",
                  showlegend=False)
```
